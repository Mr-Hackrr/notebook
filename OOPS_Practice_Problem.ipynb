````python
class Bank:
    def __init__(self):
        self.customers = {}

    def create_account(self, acc_no, initial_balance = 0):
        '''If acc_no already exists then say : Account number already exists
        For new acc_no: Enter them in the customers dictionary and say : Account created successfully'''
        if acc_no in self.customers.keys():
            print("Account number already exists")
        else:
            self.customers[acc_no] = initial_balance
            print("Account created successfully")
        

    def make_deposit(self, account_number, amount):
        '''deposit the given amount to the respective account
        For the failure case say: Account number does not exist'''
        if account_number not in self.customers.keys():
            print("Account number does not exist")
        else:
            self.customers[account_number] += amount
            print("Deposit successful")

    def make_withdrawal(self, account_number, amount):
        '''withdraw the given amount if fund exist with msg: Withdrawal successful
        else say : Insufficient funds
        For the failure case say: Account number does not exist'''
        if account_number not in self.customers.keys():
            print("Account number does not exist")
        else:
            if self.customers[account_number] >= amount:
                self.customers[account_number] -= amount
                print("Withdrawal successful")
            else:
                print("Insufficient funds")
            

    def check_balance(self, account_number):
        '''Display balance for given account
        For the failure case say: Account number does not exist'''
        if account_number not in self.customers.keys():
            print("Account number does not exist")
        else:
            print(f"Account balance : {self.customers[account_number]}")

    def make_transfer(self,src_acc, target_acc, amount):
        if src_acc not in self.customers.keys():
            print(f"Transfer declined ! {src_acc} account does not exists")
            return 
        if target_acc not in self.customers.keys():
            print(f"Transfer declined ! {target_acc} account does not exists")
            return 
        if self.customers[src_acc] >= amount:
            self.customers[src_acc] -= amount
            self.customers[target_acc] += amount
            print(f"{amount} amount transferred from {src_acc} to {target_acc}")
        else:
            print("Insufficient Funds")

if __name__ == "__main__":
    bank = Bank()
    accno1 = 'SB-123'
    damt1 = 1000
    print("New a/c No. : ",accno1,'Deposit Amount : ',damt1)
    bank.create_account(accno1,damt1)
    wamt1 = 600
    print("\nDeposit Rs.", wamt1,"to A/c No.",accno1)
    bank.make_deposit(accno1, wamt1)
    print('A/c. No.', accno1)
    bank.check_balance(accno1)
    bank.make_withdrawal(accno1,500)
    bank.check_balance(accno1)
    print()

    accno2 = 'SB-456'
    print("New a/c No. : ",accno2,'Deposit Amount : ',0)
    bank.create_account(accno2)
    bank.make_transfer(accno1, accno2, 200)
    bank.check_balance(accno1)
    bank.check_balance(accno2)

    accno3 = 'SB-159'
    accno4 = 'SB-753'
    bank.make_transfer(accno2, accno1, 500)
````
