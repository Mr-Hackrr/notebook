```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("/content/batting_stats_for_icc_mens_t20_world_cup_2024.csv")
```

```
data.head()
```

```
# 1. Top run scorer: Identify the player with the highest runs scored.
top_scorer = data.loc[data['Runs'].idxmax()]
print("Top Run Scorer:")
print(top_scorer)
```

```
#
team_performance = data.groupby("Team")["Runs"].sum()
team = team_performance.idxmax()
score = team_performance.max()
print(team)
print(score)
```

```
# 2. Highest individual score: Determine the highest score achieved by a single player in an innings.
highest_score = data['Runs'].max()
print("\nHighest Individual Score:", highest_score)
```

```
# 3. Most sixes: Find the player who hit the most sixes in the tournament.
most_sixes_player = data.loc[data['6s'].idxmax()]
print("\nMost Sixes:")
print(most_sixes_player)
```

```
data.head()
```

```
# 4. Best batting average: Calculate and display the player with the best batting average (Runs / Innings Played).
best_avg_player = data.loc[data['Ave'].idxmax()]
print("\nBest Batting Average:")
print(best_avg_player)
```

```
# 5. Highest strike rate: Identify the player with the highest strike rate (Runs per 100 balls faced).
#data['Strike Rate'] = (data['Runs'] / data['SR']) * 100
highest_sr_player = data.loc[data['SR'].idxmax()]
print("\nHighest Strike Rate:")
print(highest_sr_player)
```

```
# 6. Most runs by team: Determine the team that scored the most runs in the tournament.
team_runs = data.groupby('Team')['Runs'].sum()
highest_scoring_team = team_runs.idxmax()
print("\nHighest Scoring Team:", highest_scoring_team)
print("Total Runs:", team_runs.max())
```

```
# 7. Average runs per match: Calculate the average runs scored per match throughout the tournament.
avg_runs_per_match = data['Runs'].sum() / len(data)
print("\nAverage Runs Per Match:", avg_runs_per_match)
```

```
# 8. Distribution of runs: Visualize the distribution of runs scored by players using a histogram.
plt.hist(data['Runs'], bins=10)
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored')
plt.show()
```

```
# 9. Correlation between runs and strike rate: Investigate the relationship between runs scored and strike rate using a scatter plot.
plt.scatter(data['Runs'], data['SR'])
plt.xlabel('Runs Scored')
plt.ylabel('Strike Rate')
plt.title('Correlation between Runs and Strike Rate')
plt.show()
```

```
# 10. Top performers by team: For each team, identify the player who scored the most runs.
top_performers_by_team = data.groupby('Team')['Runs'].idxmax()
print("\nTop Performers by Team:")
for team, player_idx in top_performers_by_team.items():
    print(team + ":", data.loc[player_idx, 'Player'])
```
